```python
funcion = [f_1, f_2, f_3, f_4]
n=4
H= zeros(n, n)
S= zeros(n, n)
for i in range(n):
    for j in range(n):
        integrando = (-hbar**2/(2*m))*funcion[i]*diff(funcion[j], x, 2)
        H[i,j] = integrate(integrando, (x, 0, l))
        S[i,j] = integrate(funcion[i]*funcion[j], (x, 0, l))   
t= H - S*W 
e=t.det()
g=solve(e,W)
g
```
